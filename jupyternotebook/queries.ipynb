```python
from nsepython import equity_history, nse_fno, quote_derivative
import numpy as np
import math
```

```python
symbol = "HCLTECH"
series = "EQ"
start_date = "22-01-2024"
end_date ="21-01-2025"
```

```python
hcl = equity_history(symbol,series,start_date,end_date)
```

```python
hcl.sort_values(by='TIMESTAMP', ascending = True, inplace = True)
```

```python
hcl
```

```python
hcl.CH_CLOSING_PRICE.pct_change(1)
```

```python
hcl['dailyret'] = (np.log(hcl.CH_CLOSING_PRICE/hcl.CH_CLOSING_PRICE.shift(1)))*100
```

```python
dailyVolt_pct = hcl.dailyret.std() 
```

```python
dailyVolt_pct
```

```python
annVolt_pct = dailyVolt_pct * math.sqrt(len(hcl.dailyret))
```
