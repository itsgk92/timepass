```python
import nsepython as nse
```

```python
symbol = "HCLTECH"
series = "EQ"
start_date = "22-01-2024"
end_date ="09-02-2025"
```

```python
hcl = nse.equity_history(symbol,series,start_date,end_date)
hcl.sort_values(by='TIMESTAMP', ascending = True, inplace = True)
```

```python
hcl
```

```python
op = hcl['CH_OPENING_PRICE'].astype(float)
hi = hcl['CH_TRADE_HIGH_PRICE'].astype(float)
lo = hcl['CH_TRADE_LOW_PRICE'].astype(float)
cl = hcl['CH_CLOSING_PRICE'].astype(float)
```

```python
hcl['DOJI'] = talib.CDLDOJI(op, hi, lo, cl)
hcl['Engulfing'] = talib.CDLENGULFING(op, hi, lo, cl)
hcl['HAMMER'] = talib.CDLHAMMER(op, hi, lo, cl)
hcl['HANGINGMAN'] = talib.CDLHANGINGMAN(op, hi, lo, cl)
hcl['HARAMI'] = talib.CDLHARAMI(op, hi, lo, cl)
hcl['MORNINGSTAR'] = talib.CDLMORNINGSTAR(op, hi, lo, cl)
```

```python
hcl
```

```python
pattern_counts = hcl[['DOJI', 'Engulfing', 'HAMMER', 
                      'HARAMI', 
                      'MORNINGSTAR']].apply(lambda x: (x != 0).sum())

print(pattern_counts)
```
